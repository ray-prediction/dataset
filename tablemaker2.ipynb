```python
from graphmaker import GraphMaker
import os
import numpy as np
```

```python
scene_dir = 'manhattan_minis4'
rt_output_dir = 'manhattan_output_256'
dataset_output_dir = 'manhattan_minis4_dataset'
scene_files = [f for f in os.listdir(scene_dir) if f.endswith('.xml')]
print(scene_files)
```

```python
for scene_file in scene_files:
    print(scene_file)
    file_no_ext = os.path.splitext(scene_file)[0]
    try:
        gm = GraphMaker(os.path.join(scene_dir, scene_file))
    except:
        print('Error making the scene graph. Likely empty.')
        continue
    
    data = {}

    tx_p = np.genfromtxt(fname=os.path.join(rt_output_dir, file_no_ext, 'tx_locations'), dtype=np.float32)
    rx_p = np.genfromtxt(fname=os.path.join(rt_output_dir, file_no_ext, 'rx_locations'), dtype=np.float32)
    real_cc = np.genfromtxt(fname=os.path.join(rt_output_dir, file_no_ext, 'real_a'), dtype=np.float32)
    imag_cc = np.genfromtxt(fname=os.path.join(rt_output_dir, file_no_ext, 'imaginary_a'), dtype=np.float32)
    
    pv = gm.get_vertices().cpu().numpy()
    pn = gm.face_normals.cpu().numpy()

    env_graphs = gm.calc_graph_edges()
    env_graph_len = [len(row) for row in env_graphs]
    env_graphs = np.concatenate(env_graphs)
    
    tx_graph = gm.calc_antenna_edges(tx_p)
    tx_graph_len = [len(row) for row in tx_graph]
    tx_graph = np.concatenate(tx_graph)
    
    rx_graph = gm.calc_antenna_edges(rx_p)
    rx_graph_len = [len(row) for row in rx_graph]
    rx_graph = np.concatenate(rx_graph)
    # tx_pos = tx_p
    # rx_pos = rx_p
    # prim_vertices = pv
    # prim_normals = pn
    # real_channel_coeff = real_cc
    # imag_channel_coeff = imag_cc
    # print(env_graphs)
    
    
    output_file = file=os.path.join(dataset_output_dir, file_no_ext)
    np.savez_compressed(output_file, 
                        env_graphs=env_graphs,
                        env_graph_len=env_graph_len,
                        tx_graph=tx_graph,
                        tx_graph_len=tx_graph_len,
                        rx_graph=rx_graph,
                        rx_graph_len=rx_graph_len,
                        tx_pos=tx_p,
                        rx_pos=rx_p,
                        prim_vert=pv,
                        prim_norm=pn,
                        real_coeff=real_cc,
                        imag_coeff=imag_cc)
    
    loaded = np.load(output_file + '.npz')
    # print(np.array_equal(loaded['env_graphs'], env_graphs))
    # print(np.array_equal(loaded['tx_pos'], tx_p))
    # with open(os.path.join(dataset_output_dir, file_no_ext + '.json'), 'w') as f:
    #     json.dump(data, f, indent=4)
```
